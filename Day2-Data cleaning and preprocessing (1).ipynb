### Label Encoding

##### converts each category into a unique numerical value(labels).

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Gender']
```

##### In this method, each unique category is assigned a numerical value.
#### Example:
##### Male → 1  
##### Female → 0  
#### Purpose:
##### Machine learning models require numerical input.Label Encoding helps transform categorical data into a suitable format.
#### Observation:
##### After applying Label Encoding on the Gender column, the values were converted into integers such as 0, 1.
##### This indicates that the column contains multiple categories beyond just binary values.
#### Limitation:
##### Label Encoding introduces an ordinal relationship between categories, which may not be meaningful.
#### Conclusion:
##### Label Encoding is suitable for binary categorical variables. For columns with multiple categories, One-Hot Encoding is a better approach.

### One-Hot Encoding

##### One-Hot Encoding is a technique used to convert more categorical variables into multiple binary (0/1) columns.
#### One-Hot Encoding:
##### The "Diet" column was originally stored as integer values (0, 1, 2), representing different categories.
#### Conclusion:
##### After encoding, all features in the dataset are numerical and suitable for machine learning models.
