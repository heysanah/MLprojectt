##### Check if dataset is balanced or imbalanced.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.heatmap(df.select_dtypes(include=['number']).corr(),cmap="coolwarm")
plt.show()
```

#### Correlation Heatmap Analysis
##### A correlation heatmap was generated to understand the relationships between numerical features in the dataset.
#### Observations:
##### The diagonal values are equal to 1, indicating that each feature is perfectly correlated with itself.
##### Age vs Age = 1 , Cholesterol vs Cholesterol = 1
##### Most of the values in the heatmap are close to 0 (blue color), which indicates weak or no correlation between features.
##### There are no strong positive or negative correlations between different variables.
##### This suggests that the dataset has low multicollinearity, meaning features are not highly dependent on each other.
#### Conclusion:
##### The features in the dataset are mostly independent, which is beneficial for machine learning models. Low multicollinearity improves model stability and performance, especially for algorithms like Logistic Regression.

```python
sns.histplot(df['Age'], kde=True)
plt.show()
```

#### Age Distribution (Histogram)
##### The histogram illustrates the distribution of age among individuals in the dataset.

No extreme skewness is observed, suggesting a relatively balanced age distribution.

```python
sns.countplot(x=df[target],palette=["green","red"])
plt.title("Heart Attack Risk Distribution")
plt.show()
```

#### Heart Attack Risk Distribution (Countplot)
##### The countplot shows the distribution of individuals across the two target classes: no risk(0)and high risk(1). in X-axis class(0&1), in Y-axis count of peoples.
##### The plot indicates how many individuals belong to each category.If both classes have similar counts, the dataset is balanced.If one class dominates, the dataset is imbalanced.  Here the dataset in not balanced.
#### Insight:
##### A balanced dataset helps the model learn both classes effectively, while an imbalanced dataset may lead to biased predictions toward the majority class.
#### Conclusion:
##### Understanding class distribution is important for selecting appropriate evaluation metrics like precision & recall that help to improving model performance.

```python
import seaborn as sns
sns.boxplot(y=df['Cholesterol'])
plt.title("Cholesterol Distribution")
plt.show()
```

#### Cholesterol Distribution (Boxplot)
##### The boxplot shows the distribution of cholesterol levels among individuals.
##### middle 50% data. values approx: 0.25-0.75 range. most peoples between this range.there is no extreme high value. High cholesterol heart attack risk increase.
##### The median value represents the typical cholesterol level.The spread of the box indicates variation in cholesterol levels.Presence of outliers suggests that some individuals have unusually high cholesterol values.

```python
sns.scatterplot(x=df['Age'], y=df['Cholesterol'])
plt.title("Age vs Cholesterol")
plt.show()
```

##### The scatter plot illustrates the relationship between age and cholesterol levels.
