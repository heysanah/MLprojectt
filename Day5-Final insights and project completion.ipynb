## Day 5 -Insights & Conclusion
### Insights
#### Age and Heart Attack Risk
##### Older individuals show a higher probability of heart attack risk.
##### Visualization (histogram/boxplot) shows that high-risk patients are concentrated in higher age ranges.
#### Age is a strong predictor.
###### Cholesterol Levels Impact Risk
###### Patients with higher cholesterol levels tend to have higher risk.
###### Boxplot comparison shows clear separation between low-risk and high-risk groups.
#### Cholesterol is a key medical indicator.
##### Lifestyle Factors Matter (Smoking, Alcohol, Obesity)
##### Smoking and obesity show strong positive correlation with heart attack risk.
##### Sedentary hours also increase risk significantly.
#### Poor lifestyle habits increase risk.
##### Exercise Reduces Risk
##### Individuals with more weekly exercise hours show lower risk levels.
##### Negative relationship observed in scatter plots.
#### Exercise acts as a protective factor.
##### Stress Level is a Hidden Risk Factor
##### Higher stress levels are associated with increased heart attack probability.
#### Mental health plays a role in physical health.
##### Medical History Strongly Influences Outcome
#### Features like:
##### Previous Heart Problems,Diabetes,Family History this show strong contribution to prediction.
#### Past health conditions are critical predictors.
##### Model Performance Insight
##### Logistic Regression / Decision Tree successfully identified patterns.
##### Important features (from model):Age,Cholesterol,Exercise,Stress
#### Model confirms domain knowledge.
### Conclusion
##### This project successfully analyzed and predicted heart attack risk using machine learning techniques.
##### Key health indicators like age, cholesterol, and medical history significantly influence heart attack risk.
##### Lifestyle factors such as exercise, smoking, and stress also play a crucial role.
##### The model demonstrated the ability to classify risk effectively, making it useful for early prediction.
##### Overall, the project highlights how data analytics can support preventive healthcare decisions.
