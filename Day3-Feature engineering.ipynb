### Feature Engineering
##### New features are created from existing variables to improve model performance and uncover hidden relationships in the dataset.

```python
#Health Risk Score(Aggregation):-
df["Health_Risk_Score"]=(df["Smoking"]+df["Obesity"]+df["Diabetes"]+df["Family History"])
df["Health_Risk_Score"]
```

#### Health Risk Score
##### This feature is the sum of multiple binary risk factors:Smoking,Obesity,Diabetes,Family History.
##### Instead of treating each risk factor separately, this combines them into a single indicator of overall health risk. This can help the model detect cumulative effects.
##### Higher values of Health_Risk_Score are expected to correlate with higher heart attack risk.
##### 0 is Low risk, 4 is Very high risk

```python
#Lifestyle Balance Ratio:-
df["Lifestyle_Ratio"]=df["Exercise Hours Per Week"]/(df["Sedentary Hours Per Day"]+1)
df["Lifestyle_Ratio"]
```

##### +1 avoids division by zero.
##### This captures the balance between active and inactive lifestyle, which is more meaningful than using each variable alone.
##### Higher ratio - healthier lifestyle - lower heart attack risk.

```python
#Stress × Heart Rate:-
df["Stress_Heart_Interaction"]=df["Stress Level"]*df["Heart rate"]
df["Stress_Heart_Interaction"]
```

##### Multiplication of stress level and heart rate.
##### High stress combined with high heart rate may indicate cardiovascular strain, making this a powerful interaction feature.
###### Higher values may indicate increased heart attack risk.

### Evaluate Relationship with Target Variable

```python
target="Heart Attack Risk(Binary)"
```

```python
print(target)
print(df.columns)
```

```python
features = ["Health_Risk_Score", "Lifestyle_Ratio", "Stress_Heart_Interaction"]
target = 'Heart Attack Risk (Binary)'

for feature in features:
    print("=================================")
    print(f"{feature} vs Target:\n")
    
    result = df.groupby(target)[feature].mean()
    print(result)
```

#### Health_Risk_Score
##### 0 - 2.545 , 1 - 2.540, Difference=very very small (~0.005), Both groups (0 = no risk, 1 = risk) have almost same average score.This feature is NOT strongly separating the classes.

#### Lifestyle_Ratio
##### 0 - 0.345 , 1 - 0.350, Slightly higher for risk group (unexpected)Ideally, higher lifestyle ratio = healthier → lower risk, But here it's almost same + slightly reversed.Indicates weak or noisy relationship.

#### Stress_Heart_Interaction
##### 0 - 0.276 , 1 - 0.272, Again very small difference, No clear separation between risk groups, Interaction feature is not contributing much.
